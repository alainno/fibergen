```python
import matplotlib.pyplot as plt

img1 = cv2.imread('data/skeletons/images/0500.png', 0)
img2 = cv2.imread('data/skeletons/masks/0500.png', 0)

plt.figure(figsize=(20,10))
plt.subplot(1,2,1), plt.imshow(img1, cmap='gray')
plt.subplot(1,2,2), plt.imshow(img2, cmap='gray')
plt.show()
```

```python
start = 25
end = start + 28
img = img2[start:end,start:end]

img = np.squeeze(img)

fig = plt.figure(figsize = (12,12)) 
ax = fig.add_subplot(111)
ax.imshow(img, cmap='gray')
width, height = img.shape
thresh = img.max()/2.5
for x in range(width):
    for y in range(height):
        val = round(img[x][y],2) if img[x][y] !=0 else 0
        ax.annotate(str(val), xy=(y,x),
                    horizontalalignment='center',
                    verticalalignment='center',
                    color='white' if img[x][y]<thresh else 'black')
```
